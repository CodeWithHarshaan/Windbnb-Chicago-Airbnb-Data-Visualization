**NOTE** : Our major concern is with the price column so we have to find the major ouliers in the price column

```python
sns.boxplot(x = df_copy['price'])
plt.show()
```

**As we can see from the box plot we dont of so many outliers in price as the data is not normally distributed we can use IQR techniquue**

```python
def iqr_technique(DFcolumn):
  Q1 = np.percentile(DFcolumn, 25)
  Q3 = np.percentile(DFcolumn, 75)
  IQR = Q3 - Q1
  lower_range = Q1 - (1.5 * IQR)
  upper_range = Q3 + (1.5 * IQR)   # interquantile range

  return lower_range,upper_range
```

```python
lower_bound,upper_bound  = iqr_technique(df_copy['price'])

outliers = df_copy[(df_copy.price > lower_bound) & (df_copy.price < upper_bound)]
```
